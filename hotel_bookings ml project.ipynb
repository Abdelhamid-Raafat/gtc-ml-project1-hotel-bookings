```
df['company'] = df['company'].fillna(0).astype(int)
df['agent'] = df['agent'].fillna(0).astype(int)
```

```
df['country'] = df['country'].fillna(df['country'].mode()[0])
```

```
df = df.drop_duplicates()
```

```
df['total_guests'] = (
    df[['adults', 'children', 'babies']]
    .fillna(0)
    .sum(axis=1)
    .astype(int)
)
```

```
df['total_guests'].head(5)
```

```
df['total_nights'] = (
    df[['stays_in_weekend_nights', 'stays_in_week_nights']]
    .fillna(0)
    .sum(axis=1)
    .astype(int)
)
```

```
df['total_nights'].head(5)
```

```
df['is_family'] = (df['children'] > 0) | (df['babies'] > 0)
```
